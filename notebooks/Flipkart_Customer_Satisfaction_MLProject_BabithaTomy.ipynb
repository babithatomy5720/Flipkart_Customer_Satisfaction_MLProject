## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# STEP 1 - Import Libraries

import pandas as pd
import numpy as np
# Ignore warnings
import warnings
warnings.filterwarnings('ignore')


# STEP 2 - Upload Dataset
from google.colab import files
uploaded = files.upload()


# STEP 3 - Load Dataset
try:
    # Replace file name if needed
    df = pd.read_csv('/content/Customer_support_data.csv')
    print("Dataset Loaded Successfully")
except FileNotFoundError:
    print("File Not Found")
except Exception as e:
    print(e)



# STEP 4 - Basic Exploration
# First 5 rows
print(df.head())
# Shape of dataset
print("Dataset Shape :", df.shape)
# Column names
print(df.columns)
# Dataset information
print(df.info())


# STEP 5 - Check Duplicates
print("Duplicate Rows :", df.duplicated().sum())
# Removing duplicate rows
df.drop_duplicates(inplace=True)


# STEP 6 - Missing Values
missing_values = df.isnull().sum()
print(missing_values)


# STEP 7 - Remove Unnecessary Columns
# ID columns usually do not help prediction
cols_to_drop = ['Unique id', 'Order_id']
df.drop(columns=cols_to_drop, inplace=True)


# STEP 8 - Handle Missing Values
# Numerical Columns
num_cols = df.select_dtypes(include=np.number).columns
for col in num_cols:
    # Filling numerical missing values using median
    df[col].fillna(df[col].median(), inplace=True)

# Categorical Columns
cat_cols = df.select_dtypes(include='object').columns
for col in cat_cols:
    # Filling categorical missing values using mode
    df[col].fillna(df[col].mode()[0], inplace=True)



# STEP 9 - Datetime Conversion
df['Issue_reported at'] = pd.to_datetime( df['Issue_reported at'], dayfirst=True )
df['issue_responded'] = pd.to_datetime( df['issue_responded'], dayfirst=True )
print("Datetime Conversion Successful")


# STEP 10 - Feature Creation
# Creating response time feature
df['response_time_mins'] = (
    df['issue_responded'] - df['Issue_reported at']
).dt.total_seconds() / 60


# STEP 11 - Extract Time Features
df['issue_hour'] = df['Issue_reported at'].dt.hour
df['issue_day'] = df['Issue_reported at'].dt.day_name()


# STEP 12 - Check Unique Values
for col in df.columns:
    print(f"{col} : {df[col].nunique()} unique values")


# STEP 13 - Final Dataset Check
print(df.info())
print(df.head())


# STEP 14 - Save Cleaned Dataset
df.to_csv('cleaned_flipkart_csat_data.csv', index=False)
print("Cleaned Dataset Saved Successfully")

````

## ***5. Hypothesis Testing***

### Hypothetical Statement - 1 -> **To determine whether customer satisfaction (CSAT Score) differs across different support channels.**

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Null Hypothesis (H₀)

There is no significant difference in average CSAT scores across different support channels.

`H₀: Average CSAT scores are equal across all support channels.`

Alternate Hypothesis (H₁).

There is a significant difference in average CSAT scores across different support channels.


`H₁: At least one support channel has a different average CSAT score.`

#### 2. Perform an appropriate statistical test.

````python
df['channel_name'].unique()
````

````python
# Perform Statistical Test to obtain P-Value
# Import library

from scipy.stats import f_oneway

# Grouping CSAT scores by support channel

chat = df[df['channel_name'] == 'Outcall']['CSAT Score']

call = df[df['channel_name'] == 'Inbound']['CSAT Score']

email = df[df['channel_name'] == 'Email']['CSAT Score']

# Performing ANOVA Test

f_stat, p_value = f_oneway(chat, call, email)

# Display results

print("F-Statistic :", f_stat)

print("P-Value :", p_value)
````

````python
# Decision Making

if p_value < 0.05:

    print("Reject Null Hypothesis")

    print("There is a significant difference in CSAT scores across support channels")

else:

    print("Fail to Reject Null Hypothesis")
````

##### Which statistical test have you done to obtain P-Value?

The statistical test performed is:

One-Way ANOVA Test

`from scipy.stats import f_oneway`

The ANOVA test was used to calculate the p-value and determine whether customer satisfaction differs significantly across multiple support channels.

##### Why did you choose the specific statistical test?

The One-Way ANOVA test was chosen because:


*   the independent variable (channel_name) is categorical
*   the dependent variable (CSAT Score) is numerical


*   there are more than two support channel groups being compared

ANOVA is specifically used to compare the mean values of a numerical variable across multiple categories. In this project, it helps determine whether different support channels provide significantly different customer satisfaction levels.

````python
# Function to remove outliers using IQR

def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[
        (df[column] >= lower_bound) &
        (df[column] <= upper_bound)
    ]

# Removing outliers from selected columns
df = remove_outliers(df, 'Item_price')
df = remove_outliers(df, 'connected_handling_time')
df = remove_outliers(df, 'response_time_mins')
print("Outliers Removed Successfully")
````

````python
# =====================================================
# HANDLING IMBALANCED DATASET USING SMOTE
# =====================================================

from imblearn.over_sampling import SMOTE

# Step 1: Remove unnecessary text/NLP columns if present

cols_to_drop = [
    'Customer Remarks',
    'clean_text',
    'tokens',
    'pos_tags'
]

X_train = X_train.drop(
    columns=[col for col in cols_to_drop if col in X_train.columns]
)

X_test = X_test.drop(
    columns=[col for col in cols_to_drop if col in X_test.columns]
)

# Step 2: Keep only numerical columns

X_train = X_train.select_dtypes(include=['int64','float64'])

X_test = X_test.select_dtypes(include=['int64','float64'])

# Step 3: Handle missing values if any

X_train = X_train.fillna(0)

X_test = X_test.fillna(0)

# Step 4: Check target distribution before SMOTE

print("Before SMOTE:")
print(y_train.value_counts())

# Step 5: Apply SMOTE

smote = SMOTE(
    random_state=42
)

X_train_smote, y_train_smote = smote.fit_resample(
    X_train,
    y_train
)

# Step 6: Verify class balance

print("\nAfter SMOTE:")
print(y_train_smote.value_counts())
````
